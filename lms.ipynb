```python
def add_book():
    conn = mysql.connector.connect(
        host='localhost', user='root', password='1234')
    cursor = conn.cursor()

    title = input('Enter Book Title :')
    author = input('Enter Book Author : ')
    publisher = input('Enter Book Publisher : ')
    pages = input('Enter Book Pages : ')
    price = input('Enter Book Price : ')
    edition = input('Enter Book Edition : ')
    copies  = int(input('Enter copies : '))

    # check if the library database exists and create it if not
    cursor.execute("SHOW DATABASES")
    databases = cursor.fetchall()
    if ('library',) not in databases:
        try:
            cursor.execute("CREATE DATABASE library")
            print("Library database created successfully.")
        except mysql.connector.Error as err:
            print("Error: {}".format(err))
    else:
        conn = mysql.connector.connect(host='localhost', user='root', database='library', password='1234')
        cursor = conn.cursor()
        # create the book table if it doesn't exist
        try:
            cursor.execute('''CREATE TABLE book (
            id INT AUTO_INCREMENT PRIMARY KEY,
            title VARCHAR(255) NOT NULL,
            author VARCHAR(255) NOT NULL,
            publisher VARCHAR(255) NOT NULL,
            pages INT NOT NULL,
            price DECIMAL(10,2) NOT NULL,
            edition VARCHAR(255) NOT NULL,
            status VARCHAR(255) NOT NULL DEFAULT 'available',
            copies INT NOT NULL
        )''')
            print("Book table created successfully.")
        except mysql.connector.Error as err:
            print("Error: {}".format(err))

            # insert book data into table
            sql = '''INSERT INTO book (title, author, publisher, pages, price, edition, copies) 
             VALUES (%s, %s, %s, %s, %s, %s, %s)'''
            values = (title, author, publisher, pages, price, edition, copies)
            cursor.execute(sql, values)
            conn.commit()

            print('\n\nNew Book added successfully')
            wait = input('\n\n\n Press any key to continue....')
            conn.close()
```

```python
def add_member():
    conn = mysql.connector.connect(
    host='localhost', database='library', user='root', password='1234')
    cursor = conn.cursor()

    name = input('Enter Member Name :')
    clas = input('Enter Member Class & Section : ')
    address =input('Enter Member Address : ')
    phone = input('Enter Member Phone  : ')
    email = input('Enter Member Email  : ')

    # check if the member table exists and create it if not
    try:
        cursor.execute("SHOW TABLES LIKE 'member'")
        table_exist = cursor.fetchone()
        if table_exist is None:
            cursor.execute('''CREATE TABLE member (
        id INT AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(255) NOT NULL,
        class VARCHAR(255) NOT NULL,
        address VARCHAR(255) NOT NULL,
        phone VARCHAR(255) NOT NULL,
        email VARCHAR(255) NOT NULL
        )''')
            print("Member table created successfully.")
    except mysql.connector.Error as err:
        print("Error: {}".format(err))

        # insert member data into table
        sql = '''INSERT INTO member (name, class, address, phone, email) 
 VALUES (%s, %s, %s, %s, %s)'''
        values = (name, clas, address, phone, email)
        cursor.execute(sql, values)
        conn.commit()

        print('\n\nNew Member added successfully')
        wait =input('\n\n\n Press any key to continue....')
        conn.close()
```
